# **Regression**

Linear and polynomial

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
from sklearn import metrics
```

```python
print(data['price'].values.reshape(-1, 1).shape)
print(data['sqft_living'].values.shape)
```

```python
def reggression(data, col_x, col_y = 'price', degree = 1):
    x = data[col_x].values
    y = data[col_y].values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  #splitting data
    x_train = x_train.reshape(-1, 1)
    y_train = y_train.reshape(-1, 1)
    x_test = x_test.reshape(-1, 1)
    lg=LinearRegression()
    if degree == 1:
        
        #poly=PolynomialFeatures(degree=degree)
        
        
        #print(x_train)
        lg.fit(x_train, y_train)
        predicted = lg.predict(x_test)
    else:
        poly = PolynomialFeatures(degree = degree) 
        x_train_ = poly.fit_transform(x_train) 
        #x_train_ = poly.fit_transform(x_train) 
        lg.fit(x_train_, y_train)
        x_test_ = poly.fit_transform(x_test)
        predicted = lg.predict(x_test_)
    
    plot_scatter(data,x_test, y_test,col_x,col_y)
    plt.plot(x_test, predicted)
    print("MSE: ", metrics.mean_squared_error(y_test, predicted))
    print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
reggression(data, 'sqft_living')
```

```python
reggression(data, 'sqft_living', degree=2)
#data.sqft_living.values.reshape(1,-1)
```

```python
reggression(data, 'sqft_living', degree=3)
```

```python
reggression(data, 'sqft_living', degree=4)
```

from the above exemples we note that the best model is the poly regression (degree = 2), because it has the smallest MSE

# **Multi Regression**

```python
def multi_reggression(data, cols_x, col_y = 'price', degree = 1):
    all_features = [data[c].values for c in cols_x]
    x = []
    for e in zip(*all_features):
        x.append(list(e))
    
    y = data[col_y].values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=40)  #splitting data
    #x_train = x_train.reshape(-1, 1)
    #y_train = y_train.reshape(-1, 1)
    #x_test = x_test.reshape(-1, 1)
    lg=LinearRegression()
    if degree == 1:
        
        #poly=PolynomialFeatures(degree=degree)
        
        
        #print(x_train)
        lg.fit(x_train, y_train)
        predicted = lg.predict(x_test)
    else:
        poly = PolynomialFeatures(degree = degree) 
        x_train_ = poly.fit_transform(x_train) 
        #x_train_ = poly.fit_transform(x_train) 
        lg.fit(x_train_, y_train)
        x_test_ = poly.fit_transform(x_test)
        predicted = lg.predict(x_test_)
    
    #plot_scatter(data,x_test, y_test,col_x,col_y)
    #plt.plot(x_test, predicted)
    print("MSE: ", metrics.mean_squared_error(y_test, predicted))
    print("R squared: ", metrics.r2_score(y_test,predicted))
```

```python
multi_reggression(data, ['bathrooms','sqft_living' ], degree=1)
```

```python
multi_reggression(data, ['bathrooms','sqft_living' ], degree=2)
```

```python
multi_reggression(data, ['bathrooms','sqft_living' ], degree=3)
```

```python
multi_reggression(data, ['bathrooms','sqft_living' ], degree=4)
```

```python
multi_reggression(data, ['bathrooms','sqft_living', 'bedrooms' ], degree=1)
```

```python
multi_reggression(data, ['bathrooms','sqft_living', 'bedrooms' ], degree=2)
```

```python
multi_reggression(data, ['bathrooms','sqft_living', 'bedrooms' ], degree=3)
```

in the above exemples the best result was obtained when using 2 features ('bathrooms','sqft_living') with a poly regression (degree = 2)
