In data science, preprocessing refers to the steps taken to clean, transform, and prepare raw data for analysis. Preprocessing is a critical step in the data analysis process, as it can significantly impact the accuracy and effectiveness of the final results.

Here are some common preprocessing steps in data science:

**Data Cleaning:** This involves removing or correcting data that is incomplete, inconsistent, or inaccurate. Common tasks in data cleaning include handling missing values, identifying and correcting outliers, and correcting formatting errors.

**Data Transformation**: This involves converting raw data into a format that is easier to work with or analyze. Common tasks in data transformation include normalization, scaling, and encoding categorical variables.

**Feature Extraction**: This involves selecting or creating features (i.e., variables) that are relevant to the problem at hand. Feature extraction is often used in natural language processing and computer vision applications.

**Feature Selection**: This involves selecting the most important features for a given problem. Feature selection can help reduce the dimensionality of the data and improve the performance of machine learning models.

**Data Integration**: This involves combining data from multiple sources into a single dataset. Data integration can help provide a more comprehensive view of the problem being analyzed.

# **Missing values**

**Is this value missing because it wasn't recorded or because it doesn't exist?**

If a value is missing becuase it doesn't exist (like the height of the oldest child of someone who doesn't have any children) then it doesn't make sense to try and guess what it might be. These values you probably do want to keep as NaN. On the other hand, if a value is missing because it wasn't recorded, then you can try to guess what it might have been based on the other values in that column and row. This is called imputation

Check for missing values: Use the isnull() method to identify any missing values in the dataframe, and then decide on how to handle them.

 **Scaling**

Scaling means that you're transforming your data so that it fits within a specific scale, like 0-100 or 0-1. You want to scale data when you're using methods based on measures of how far apart data points are

By scaling your variables, you can help compare different variables on equal footing

**Why do we need scaling?**

Machine learning algorithm just sees number — if there is a vast difference in the range say few ranging in thousands and few ranging in the tens, and it makes the underlying assumption that higher ranging numbers have superiority of some sort

There are many types of scaling, choosing the rigth one depends on the meaning of the features

***Max-min scaling***

Max-min scaling rescales the data to a range between 0 and 1, based on the maximum and minimum values in the dataset. This can be done using the following formula:

x_scaled = (x - min(x)) / (max(x) - min(x))

**Linear Scaling**

In our case, we can consider using a simple linear scaling method to bring the grades from the two different unit systems onto the same scale

To convert the 20-max system to a 10-max system, you can divide each grade by 2.

We can also convert the 10-max system to 20-max system by multiplying each grade (10-max system) by 2.

Since we have only 1 feature that follows the 10-max system, it is easier to multiply by 2.

Keep in mind that scaling grades in this way does not change the relative performance of the students within each system, but it does allow for easier comparison between the two systems. Additionally, scaling should be done consistently and transparently throughout your analysis to ensure the accuracy and validity of your results.
