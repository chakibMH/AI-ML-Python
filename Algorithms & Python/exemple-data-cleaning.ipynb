# I Have papers data that I need to clean the abstract, but some data is duplicate and other is useless...
# 
In this notebook I will show how to select only data I am intrested in. In my case I don't need data that exist already, neither duplicate data.

```python
import pandas as pd
import os
```

we select the files we are intrested in

```python
p = "../input/acm-papers-authors-dataset"
list_files = os.listdir(p)
```

create a list containing dataframe of the precedent files

```python
frames = []
for fn in list_files:
    frames.append(pd.read_csv(p+'/'+fn))
```

concat the frames, and that's it !

```python
result = pd.concat(frames)
result.shape
```

Eliminate rows with 'Exists' in title

First select the indexes to drop

```python
idx_to_drop = result.loc[result.title == 'Exists'].index
```

Then call the drop method inlpace

```python
result.drop(idx_to_drop, inplace = True)
```

```python
result.shape
```

remove duplicates

```python
result.drop_duplicates('id_paper', inplace = True)
```

```python
result.shape
```

save the results

```python
result.to_csv("ACM_scr_2.csv", index=False)
```
