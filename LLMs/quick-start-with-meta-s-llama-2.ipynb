# Initiating The Model

- We first need to add the model that you want to test (depending on the model version, this will take much or less time)
- Once the model is loaded, it will apperars in the Models panel at the rigth
- Then we will copy the path to the model we just added
- Finaly, Instantiate the tokernizer one time, and use the variable as a paramaterin for the following function

**DO NOT FORGET TO TURN THE GPU ON**

```python
model = '/kaggle/input/llama-2/pytorch/7b-chat-hf/1'
model = '/kaggle/input/llama-2/pytorch/13b-chat-hf/1'

tokenizer_13bChat = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
def run_Llma2(tokenizer, prompt):
    sequences = pipeline(
            prompt,
           do_sample=True,
            top_k=10,
            num_return_sequences=1,
            eos_token_id=tokenizer.eos_token_id,
            max_length=500,
        )
    
    return sequences[0]['generated_text'].replace(prompt, "")
```
